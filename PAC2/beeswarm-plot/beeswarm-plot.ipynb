# Beeswarm Plot

Per a visualitzar aquest plot utilitzarem les cotitzacions de Twitter a la borsa: https://www.kaggle.com/datasets/amandam1/twitter-daily-stock-prices-20132022



```python
import pandas as pd

df = pd.read_csv('TWTR.csv')

```

```python
df[['Year', 'Month', 'Day']] = df['Date'].str.split('-',2, expand=True).astype(int)

df = df[['High', 'Year', 'Month', 'Day']]

df.head()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize=(10,5))
ax.set_title("Valor bursatil de Twitter. Màxim diari.")

plot = sns.swarmplot(y='High', data=df, ax=ax, )
plot.set_ylabel('Valor ($ per Acció)')
plt.box(False)
```

```python
fig, ax = plt.subplots(figsize=(25,5))
ax.set_title("Valor bursatil de Twitter. Màxim diari.")

plot = sns.swarmplot(y='High', x="Year", data=df, ax=ax, )
plot.set_ylabel('Valor ($ per Acció)')
plot.set_xlabel('Any')
plt.box(False)
```
