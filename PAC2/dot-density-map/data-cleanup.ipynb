```python
# LLegir el fitxer
import pandas as pd

data = pd.read_csv('data.csv')

print(data.shape)
display(data.head())
```

```python
data = data[['Municipi', 'Latitud', 'Longitud', 'Activitat principal']]
set(data['Activitat principal'])
lluita = [ 'Aikido',
 'Boxa',
 'Capoeira',
 'Defensa personal',
 'Full contact',
 'Jiu-jitsu',
 'Judo',
 'Karate',
 'Kempo karate',
 'Kick-boxing',
 'Kung-fu',
 'Lluita grecoromana',
 'Lluita lliure olimpica',
 'Muay-thai',
 'Nihon-tai-jitsu',
 'Ta-ka-ta',
 'Taekwondo',
 'Tai-jitsu',
 'Wu-shu']
equip =[ 'Basquet',
 'Basquet lliure 3x3',
 'Beisbol',
 'Futbol',
 'Futbol america',
 'Futbol platja',
 'Futbol sala',
 'Futbol-7',
 'Handbol',
 'Hoquei herba',
 'Hoquei sala',
 'Hoquei sobre patins',
 'Hoquei sobre patins en linia',
 'Hoquei-gel',
 'Mini volei',
 'Minibasquet',
 'Rugbi',
 'Rugbi a 7',
 'Softbol',
 'Volei platja',
 'Voleibol',
 'Waterpolo']
muntanya =[ 'Alpinisme',
 'Curses de muntanya',
 'Escalada esportiva',
 'Escalades en roca',
 "Esports d'hivern (altres)",
 'Esqui alpi',
 'Esqui de fons',
 'Esqui de travessia o de muntanya',
 'Surf de neu (snowboard)',
 'Raquetes a neu']
fitness = [ 'Aerobic',
 'Cardiovasculars (activitats en cinta, eliptica, rem,...)',
 'Ciclisme en sala (spinning)',
 'Condicionament fisic (activitats dirigides)',
 'Correr (footing)',
 'Esport acrobatic',
 'Fisic-culturisme',
 'Gimnastica artistica',
 'Gimnastica ritmica',
 'Gimnastica- trampoli',
 'Ioga',
 'Esquaix',
 'Hidromassatge',
 'Musculacio i tonificacio']
aquatiques = ['Activitats aquatiques dirigides (aiguagim i similars)',
 'Altres activitats subaquatiques',
 'Caiac de mar',
 'Caiac polo',
 'Catamarans',
 'Escafandrisme',
 'Motonautica',
 'Motos aquatiques',
 'Natacio',
 'Natacio ensenyament',
 'Natacio recreativa',
 'Natacio sincronitzada',
 'Orientacio subaquatica',
 'Pati catala de vela',
 'Pesca continental',
 'Pesca maritima',
 'Pesca submarina',
 'Piraguisme aigues braves',
 'Piraguisme aigues tranquil?les',
 'Surf',
 'Surf de vela (windsurf)',
 'Vela creuers',
 'Vela lleugera (no olimpica)',
 'Vela lleugera (olimpica)',
 'Barques pneumatiques',
 'Descens aigues braves',
 'Descens en bot (rafting)',
 'Esqui nautic',
 'Salts de natacio']

def get_category(row):
    if row['Activitat principal'] in lluita:
        return 'lluita'
    if row['Activitat principal'] in equip:
        return 'equip'
    if row['Activitat principal'] in muntanya:
        return 'muntanya'
    if row['Activitat principal'] in aquatiques:
        return 'aquatiques'
    if row['Activitat principal'] in fitness:
        return 'fitness'
    return 'altres'
data['Categoria'] = data.apply(get_category, axis=1)
data = data.rename(columns={'Activitat principal':'Activitat'})
data = data[data.Categoria != 'altres']
data.head()
```

```python
data.to_csv('data-clean.csv', index_label='Id')
```
