# Icon Chart

Aquest fitxer mostra un exemple de visualitzaci√≥ de Icon Chart.

Les dades escollides per a demostrar-ho corresponen als resultats de les eleccions al Congreso de los Diputados a Catalunya al 2016. Aquestes dades han estat extretes de https://analisi.transparenciacatalunya.cat/Demografia/Eleccions-al-Congr-s-dels-Diputats/3zh5-pq3t

```python
# Per motius de simplicitat s'han copiat les dades aqui directament. 
resultats = {'ECP': 853102, 'ERC': 632234, 'PSC': 559870, 'CDC': 483488, 'PP': 464538, 'Cs': 380497, 'Altres': 74897}

import pandas as pd
import numpy as np
df = pd.Series(resultats)
# DataFrame en percentatje de vot
df = df / df.sum() * 100


# colors dels partit politics
colors=['darkviolet', 'yellow', 'red', 'turquoise', 'blue', 'orange', 'lightgrey']
```

```python
from pywaffle import Waffle
import matplotlib.pyplot as plt

labels = [partit+ " - " + "{:.2f}%".format(df[partit]) for partit in df.index]

fig = plt.figure(FigureClass=Waffle,vertical = True, figsize=(7,7), starting_location='NW', 
                 values=df.to_dict(), colors=colors, columns=10, icons='thumbs-up', 
                 block_arranging_style='new-line',
                labels=labels, legend={'bbox_to_anchor': (1.55, 1), 'fontsize':16, 'frameon': False})
plt.title('Resultats a Catalunya de les elecccions al Congreso de los Diputados. 2016', fontsize=20)
plt.show()
```
